## Introduction

There are many ways to use deep learning neural networks to decode and classify EEG. The more common training data (input to the network) are EEG features after morlet wavelets, images generated using raw EEG or EEG features, raw EEG signals, etc. . 

As for the network structure, in addition to the common CNN, DNN, LSTM, there are also self-supervised AE/VAE, and some Transformer-based models, etc.
